# Tycho SN1572

```python
d = np.linspace(1, 500, 100)
MV = [0, -3, -3.2, -4.0, -2.4, -1.7, -1.4, 0.3, 1.6, 2.5, 4.0, 4.2, 4.7, 5.3]
MV_err = [0, 0.3, 0.3, 0.3, 0.3, 0, 0.4, 0.2, 0.5, 0.2, 0.2, 0.2, 0.2, 0.4]
dates = np.array([-11, -10, -9.9, -5, 24, 47, 55, 101, 161, 253, 345, 359, 406, 451]) + 11
dates_err = [0, 0, 0, 1, 7, 0, 7, 14, 14, 14, 14, 7, 14, 7]
```

```python
fig, ax = plt.subplots()
ax.errorbar(x=dates, y=MV, xerr=dates_err, yerr=MV_err, marker='o', capsize=3)
ax.invert_yaxis()
```

# Add some points

```python
d = np.linspace(1, 500, 100)
MV = [0, -3, -3.2, -4.0, -3.6, -2.4, -1.7, -1.4, 0.3, 1.6, 2.5, 4.0, 4.2, 4.7, 5.3]
MV_err = [0, 0.3, 0.3, 0.3, 0, 0.3, 0, 0.4, 0.2, 0.5, 0.2, 0.2, 0.2, 0.2, 0.4]
dates = np.array([-11, -10, -9.9, -5, 5, 24, 47, 55, 101, 161, 253, 345, 359, 406, 451]) + 11
dates_err = [0, 0, 0, 1,0, 7, 0, 7, 14, 14, 14, 14, 7, 14, 7]

fig, ax = plt.subplots()
ax.errorbar(x=dates, y=MV, xerr=dates_err, yerr=MV_err, marker='o', capsize=3)
ax.invert_yaxis()
```

# Just a linear fit

```python
from scipy.interpolate import CubicSpline, interp1d
cs = interp1d(dates, MV, kind='slinear')
xs = np.arange(min(dates), max(dates), 1)

fig, ax = plt.subplots()
ax.errorbar(x=dates, y=MV, xerr=dates_err, yerr=MV_err, marker='o', capsize=3, linestyle='none')
# ax.plot(x_fit, y_fit, '-', label='Fitted curve')
ax.plot(xs, cs(xs), label="S")
ax.invert_yaxis()
```
